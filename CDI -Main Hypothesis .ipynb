```python
#encoding Categorical Features:
from sklearn import preprocessing
le= preprocessing.LabelEncoder()
categorical=["country","gender","Diarrhea"]
for c in categorical:
    df[c]=le.fit_transform(df[c])
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,roc_auc_score
from sklearn.linear_model import LogisticRegression
```

```python
target=df["Diarrhea"]
features=df.drop("Diarrhea", axis=1)
features.head()
```

```python
#Using Random Forest classifier on Un-balanced data:

def RFModel(X,y):
    X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0)
    rf=RandomForestClassifier()
    rf.fit(X_train,y_train)
    y_predict=rf.predict(X_test)
    print(confusion_matrix(y_test,y_predict))
    print(classification_report(y_test,y_predict))
    print("Accuracy score:",accuracy_score(y_test, y_predict))
RFModel(features,target)
```

-The reason behind the inaccurace results is the imbalance in data. 

```python
#Balancing the data using SMOTE technique:
from imblearn.over_sampling import SMOTE
X_resampled, y_resampled = SMOTE().fit_resample(features,target)
```

```python
def calculate_tpr_fpr(y_test, y_pred):
       
    # Calculates the confusion matrix and recover each element
    cm = confusion_matrix(y_test, y_pred)
    TN = cm[0, 0]
    FP = cm[0, 1]
    FN = cm[1, 0]
    TP = cm[1, 1]
    
    # Calculates tpr and fpr
    tpr =  TP/(TP + FN) # sensitivity - true positive rate
    fpr = 1 - TN/(TN+FP) # 1-specificity - false positive rate
    
    return tpr, fpr
```

```python
def get_all_roc_coordinates(y_test, y_proba):
    tpr_list = [0]
    fpr_list = [0]
    for i in range(len(y_proba)):
        threshold = y_proba[i]
        y_pred = y_proba >= threshold
        tpr, fpr = calculate_tpr_fpr(y_test, y_pred)
        tpr_list.append(tpr)
        fpr_list.append(fpr)
    return tpr_list, fpr_list
```

```python
#Getting ROC AUC score using Random forest classifier:

def RFModel_OVR(X,y):
    X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0)
    rf=RandomForestClassifier()
    rf.fit(X_train,y_train)
    y_proba=rf.predict_proba(X_test)
    classes=rf.classes_
    roc_auc_ovr = {}
    for i in range(len(classes)):
        c = classes[i]

        # Prepares an auxiliary dataframe to help with the plots
        df_aux = X_test.copy()
        df_aux['class'] = [1 if y == c else 0 for y in y_test]
        df_aux['prob'] = y_proba[:, i]
        df_aux = df_aux.reset_index(drop = True)
        tpr, fpr = get_all_roc_coordinates(df_aux['class'], df_aux['prob'])
        roc_auc_ovr[c] = roc_auc_score(df_aux['class'], df_aux['prob'])
    avg_roc_auc = 0
    i = 0
    for k in roc_auc_ovr:
        avg_roc_auc += roc_auc_ovr[k]
        i += 1
    print(f"average ROC AUC OvR: {avg_roc_auc/i:.4f}")

RFModel_OVR(X_resampled,y_resampled)
```

```python
#Applying Random Forest classifier on balanced data:
def RFModel(X,y):
    
    X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0)
    rf=RandomForestClassifier()
    rf.fit(X_train,y_train)
    y_predict=rf.predict(X_test)
    print(confusion_matrix(y_test,y_predict))
    print(classification_report(y_test,y_predict))
    print("Accuracy score:",accuracy_score(y_test, y_predict))
    print("MCC:",metrics.matthews_corrcoef(y_test, y_predict))

RFModel(X_resampled,y_resampled)
```
