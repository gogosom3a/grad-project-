```python
df=df.dropna()
```

```python
from sklearn import preprocessing
le=label_encoder = preprocessing.LabelEncoder()
df["class"]=le.fit_transform(df["class"])
```

```python
df["class"]
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,roc_auc_score
from sklearn.linear_model import LogisticRegression
```

```python
target=df["class"]
features=df.drop("class", axis=1)
features.head()
```

```python
#Applying Random Forest Classifier on Un_balanced data:

def RFModel(X,y):
    
    
    X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0)
    rf=RandomForestClassifier()
    rf.fit(X_train,y_train)
    y_predict=rf.predict(X_test)
    print(confusion_matrix(y_test,y_predict))
    print(classification_report(y_test,y_predict))
    print("Accuracy score:",accuracy_score(y_test, y_predict))
   

RFModel(features,target)
```

-The numbers generated above are inconsistent.This is because the model only predicted the majority classes.So, the data must be balanced. 

```python
#balancing data using SMOTE anaysis:
from imblearn.over_sampling import SMOTE
X_resampled, y_resampled = SMOTE().fit_resample(features,target)
```

```python
#Applying Random forest Classifier on balanced data:

RFModel(X_resampled,y_resampled)
```

```python
#Applying Gradient Boosting Classifier on balanced data:
def GBModel(X,y):
    
    
    X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0)
    rf=GradientBoostingClassifier()
    rf.fit(X_train,y_train)
    y_predict=rf.predict(X_test)
    print(confusion_matrix(y_test,y_predict))
    print(classification_report(y_test,y_predict))
    print("Accuracy score:",accuracy_score(y_test, y_predict))
    
GBModel(X_resampled,y_resampled)
```

```python
#Testing the K Nearest Neighbors classifier Model using new data:
def KNNModel(X,y):
    
    X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0)
    rf=KNeighborsClassifier()
    rf.fit(X_train,y_train)
    y_predict=rf.predict(X_test)
    print(confusion_matrix(y_test,y_predict))
    print(classification_report(y_test,y_predict))
    print("Accuracy score:",accuracy_score(y_test, y_predict)) 
KNNModel(X_resampled,y_resampled)
```
